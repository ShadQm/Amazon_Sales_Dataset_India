## Product Size Distribution

```python
#Exploratory Data Analysis
ax = sns.countplot(x='Size',hue='Size', data= df, palette='viridis', legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Product Size Distribution", fontsize=14)


#Note
plt.text(
    x=0.5, y=-0.15,
    s="Note: Data based on the latest Amazon sales report",
    ha='center', va='center', transform=ax.transAxes,
    fontsize=10, color='black'
)
plt.tight_layout()
plt.show()
```

```python
# Grouping by size and sorting by total quantity
grouped_quantity = df.groupby(['Size'], as_index =False)['Qty'].sum().sort_values(by='Qty', ascending= False)
grouped_quantity
```

## Quantity By Size

```python
quantity_graph = sns.barplot(x='Size', y= 'Qty', data= grouped_quantity,hue='Size', palette ='viridis')
plt.title('Total Quantity By Size')

for bars in quantity_graph.containers:
    quantity_graph.bar_label(bars)
plt.show()
```

## Courier Status

```python
plt.figure(figsize=(10,5))
courier_graph = sns.countplot(x='Courier Status', data= df, hue= 'Status')
plt.ylabel = ("Total Count")
plt.xlabel = ("Courier Status")
for bars in courier_graph.containers:
    courier_graph.bar_label(bars, fontsize=6)
plt.show()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
category_histogram = plt.hist(column_data, bins=20, edgecolor ='Black')
plt.xticks(rotation= 90)

plt.show()
```

```python

# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state', palette='viridis', hue='ship-state')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# top_10_States 
top_10_state = df['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12, 6))
state_plot = sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state', hue='ship-state', palette='viridis')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
for bars in state_plot.containers:
    state_plot.bar_label(bars)
plt.show()
```

```python
# Top 10 states
top_10_state = df['ship-state'].value_counts().head(10).sort_values(ascending=True)

# Generate colors from the viridis colormap
colors = plt.cm.viridis(np.linspace(1, 0, len(top_10_state)))

plt.figure(figsize=(10,6))
bars = plt.barh(top_10_state.index, top_10_state.values, color=colors, edgecolor='black')
for bar in bars:
    width = bar.get_width()
    plt.text(width + 50, bar.get_y() + bar.get_height()/2, str(width), va='center', fontsize=8)
plt.title('Top 10 States by Orders')
plt.show()
```
